# Segmenting and Clustering Neighborhoods in Toronto

## _Webpage scraping using Beautifulsoup and put the resut in a list T2_

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(source, 'lxml')
table = soup.find('table')
T2=[]
for row in table.find_all('td') :
    T2.append(row.get_text())
```

## _Convert the list T2 to Dataframe DF_

```python
n = 3
final = [T2[i * n:(i + 1) * n] for i in range((len(T2) + n - 1) // n )]  
DF=pd.DataFrame(final,columns=['Postalcode','Borough','Neighborhood'])
DF
```

## _Drop cells with a borough that is Not assigned_

```python
DF.drop(DF.loc[DF['Borough']=='Not assigned'].index,inplace=True)
DF
```

## _Combination of rows having the same neighborhoods_

```python
g=0
while g < DF.shape[0]-1 :
            if  DF.Postalcode.iloc[g+1] == DF.Postalcode.iloc[g] :
                DF.replace(to_replace = DF.Neighborhood.iloc[g],value =DF.Neighborhood.iloc[g]+','+ DF.Neighborhood.iloc[g+1] ,inplace =True)
                DF.drop(DF.index[g+1],inplace=True)
                g=0
                continue
            else:
                g+=1
DF.reset_index(drop=True,inplace = True)
DF
```

## _Assign the Brorough's name to the Neighborhood's name   when the Neighborhood's name has 'not assigned' as value_ 

```python
s=0
while s < DF.shape[0]-1 :
            if  DF.Neighborhood.iloc[s]=='Not assigned\n' :
                DF.replace(to_replace= DF.Neighborhood.iloc[s] ,value = DF.Borough.iloc[s], inplace = True)
            else:
                s+=1
DF
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
        
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)         
        results = requests.get(url).json()["response"]['groups'][0]['items']
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```
